지금까지는 데이터를 머신러닝 모델에 넣기 위한, 데이터 전처리 과정이었습니다.

뒤에 이어질 내용은 전처리 된 데이터를 머신러닝 모델에 적용시키는 부분입니다.

이는 머신러닝에 대한 추가적인 지식이 있어야 이해가 가능할 것이라고 생각됩니다.

이번 시간은 여기까지가 끝입니다. 다만, 위에서 가공된 데이터가 어떤식으로 머신러닝에 사용되는지 간단한 예시를 보여드리기 위함이므로,

`아래 내용은 선택적으로 보시길 바라겠습니다.`

```python
# test data set load
test = pd.read_csv('./타이타닉/test.csv')

print(test.shape)
test.head()
```

test 데이터에도 train 데이터에 한 전처리 과정을 똑같이 해줍니다.

```python
import re

def classify_identity(name):
    title = re.search(r'([A-za-z]+)\.', name).group() # 정규표현식을 이용해 '문자.' 형태인 문자열을 추출해냅니다.
    return title

# write your code here !!
test['Title'] = test['Name'].apply(classify_identity)

test['Title'] = test['Title'].replace(['Dr.', 'Rev.', 'Col.', 'Mlle.','Major.', 'Mme.', 'Sir.', 'Ms.', 'Capt.', 'Don.', 'Jonkheer.','Countess.', 'Lady.'], "Other")
test['Title'].value_counts()

test['Sex'] = test['Sex'].replace('male', 1).replace('female', 0)
test.drop(columns=['Ticket', 'Cabin'], axis = 1, inplace=True)

test['Family'] = test['SibSp'] + test['Parch']
test.drop(columns=['SibSp', 'Parch'], inplace=True)
test['Age'].fillna(test['Age'].mean(), inplace = True)
test['Age_clean'] = pd.cut(test['Age'], 5)

def age_cleaning(age):
    if age <= 16:
        return "Child"
    elif age <=32:
        return "Young"
    elif age <=48:
        return "Middle"
    elif age <= 64:
        return "Prime"
    else:
        return "Senior"
    
test['Age_clean'] = test['Age'].apply(age_cleaning)
train['Embarked'].fillna('S', inplace=True)
```

```python
test.head()
```

머신러닝 모델에 넣을 특징은 다음과 같습니다.

Pclass, Sex, Embarked, Title, Family, Age_clean

```python
features = ['Pclass', 'Sex', 'Embarked', 'Title', 'Family', 'Age_clean']

# 데이터의 레이블
train_label = train['Survived']

# 모델에 넣을 특징만 남긴다.
train_data = train[features]
test_data = test[features]
```

```python
train_data.head()
```

위에서 보시다시피, Embarked, Title, Age_clean은 숫자가 아닙니다.

`머신러닝 알고리즘은 숫자만 인식할 수 있기 때문에 One-Hot-Encoding을 통해서 숫자로 바꿔보겠습니다.`

[원핫인코딩]https://programmers.co.kr/learn/courses/21/lessons/11043

```python
train_data = pd.get_dummies(train_data, columns=['Embarked', 'Title', 'Age_clean'])
test_data = pd.get_dummies(test_data, columns=['Embarked', 'Title', 'Age_clean'])
```

자, 그럼 다시 숫자만 가지고 features를 정의해보겠습니다.

```python
features = ['Pclass', 'Sex', 'Family',
            'Embarked_C', 'Embarked_Q', 'Embarked_S',
           'Title_Master.', 'Title_Miss.', 'Title_Mr.', 'Title_Mrs.', 'Title_Other', 'Age_clean_Child',
            'Age_clean_Middle', 'Age_clean_Prime', 'Age_clean_Senior', 'Age_clean_Young']

train_data = train_data[features]
test_data = test_data[features]

train_data.head()
```

# Machine Learning Modeling

```python
from sklearn.ensemble import RandomForestClassifier # 사이킷런은 각종 머신러닝 알고리즘을 편리하게 사용할 수 있도록 도와줍니다.
```

```python
model = RandomForestClassifier() # 랜덤포레스트 모델 객체를 생성
```

```python
model.fit(train_data, train_label) # train_data와 label을 모델에 넣어서 학습시킨다.
```

```python
prediction = model.predict(test_data) # test 데이터를 넣었을때 모델의 예측값을 prediction에 넣는다.
```

```python
accuracy = round(model.score(train_data, train_label)*100, 2)
# score(features, label)을 사용해서 모델이 예측한 값과 실제 값을 비교해서 정확도를 측정한다.
print(accuracy)
```
