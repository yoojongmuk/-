# Pandas apply, 파이썬 함수를 적용할 수 있게 하는 함수

이것을 이용해서, 퍼스트 클래스에 탄 사람이라면 모두 VIP로 바꿔보자.

````python
def is_vip(row):
    fare = row["Fare"]
    pclass = row["Pclass"]
    
    if fare>200 and pclass == "퍼스트":
        return True
    else:
        return False

train["VIP"] = train.apply(is_vip, axis=1)
````
